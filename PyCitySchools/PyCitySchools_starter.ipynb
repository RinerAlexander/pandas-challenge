```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
sd_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
schoolList=[]
schoolDataBySchoolName=school_data.set_index("school_name")
for school in school_data["school_name"]:
    schoolList.append({"School Name":school,
                       "School Type":schoolDataBySchoolName.loc[school,"type"],
                       "Total Students":sd_complete[sd_complete["school_name"]==school].shape[0],
                       "School Budget":schoolDataBySchoolName.loc[school,"budget"],
                       "Per Student Budget":schoolDataBySchoolName.loc[school,"budget"]/sd_complete[sd_complete["school_name"]==school].shape[0],
                       "Average Math Score":sd_complete[sd_complete["school_name"]==school]["math_score"].mean(),
                       "Average Reading Score":sd_complete[sd_complete["school_name"]==school]["reading_score"].mean(),
                       "% Passing Math":sd_complete[(sd_complete["school_name"]==school)&(sd_complete["math_score"]>=70)].shape[0]/sd_complete[sd_complete["school_name"]==school].shape[0]*100,
                       "% Passing Reading":sd_complete[(sd_complete["school_name"]==school)&(sd_complete["reading_score"]>=70)].shape[0]/sd_complete[sd_complete["school_name"]==school].shape[0]*100,
                       "% Overall Passing":sd_complete[(sd_complete["school_name"]==school)&((sd_complete["reading_score"]>=70)&(sd_complete["math_score"]>=70))].shape[0]/sd_complete[sd_complete["school_name"]==school].shape[0]*100
                      })
schoolDF=pd.DataFrame(schoolList)
schoolDF
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninthdf=sd_complete[sd_complete["grade"]=="9th"]
ninthdf=ninthdf[["school_name","math_score"]]
ninthdf=ninthdf.groupby(["school_name"])
ninthdf=ninthdf.mean()
ninthdf=ninthdf.rename(columns={"math_score":"9th"})

tenthdf=sd_complete[sd_complete["grade"]=="10th"]
tenthdf=tenthdf[["school_name","math_score"]]
tenthdf=tenthdf.groupby(["school_name"])
tenthdf=tenthdf.mean()
tenthdf=tenthdf.rename(columns={"math_score":"10th"})

MathByGrade=pd.merge(ninthdf,tenthdf,on="school_name")

eleventhdf=sd_complete[sd_complete["grade"]=="11th"]
eleventhdf=eleventhdf[["school_name","math_score"]]
eleventhdf=eleventhdf.groupby(["school_name"])
eleventhdf=eleventhdf.mean()
eleventhdf=eleventhdf.rename(columns={"math_score":"11th"})

MathByGrade=pd.merge(MathByGrade,eleventhdf,on="school_name")

twelthdf=sd_complete[sd_complete["grade"]=="12th"]
twelthdf=twelthdf[["school_name","math_score"]]
twelthdf=twelthdf.groupby(["school_name"])
twelthdf=twelthdf.mean()
twelthdf=twelthdf.rename(columns={"math_score":"12th"})

MathByGrade=pd.merge(MathByGrade,twelthdf,on="school_name")


MathByGrade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
firstQ=schoolDF["Per Student Budget"].quantile(.25)
secondQ=schoolDF["Per Student Budget"].quantile(.50)
thirdQ=schoolDF["Per Student Budget"].quantile(.75)
fourthQ=schoolDF["Per Student Budget"].quantile(1)

bins=[0,firstQ,secondQ,thirdQ,fourthQ]
names=[f"first quartile {firstQ} and lower",f"Second quartile {secondQ} to {firstQ}",f"Third quartile {thirdQ} to {secondQ}",
       f"Fourth quaritle {fourthQ} to {thirdQ}"]
SchoolSpending=schoolDF
SchoolSpending["Spending Quartiles"]=pd.cut(SchoolSpending["Per Student Budget"],bins,labels=names)
SchoolSpending=SchoolSpending.groupby(["Spending Quartiles"])
SchoolSpending=SchoolSpending.mean()
SchoolSpending=SchoolSpending[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
SchoolSpending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
firstQ=schoolDF["Total Students"].quantile(.25)
secondQ=schoolDF["Total Students"].quantile(.50)
thirdQ=schoolDF["Total Students"].quantile(.75)
fourthQ=schoolDF["Total Students"].quantile(1)

bins=[0,firstQ,secondQ,thirdQ,fourthQ]
names=[f"first quartile {firstQ} and lower",f"Second quartile {secondQ} to {firstQ}",f"Third quartile {thirdQ} to {secondQ}",
       f"Fourth quaritle {fourthQ} to {thirdQ}"]
SchoolSize=schoolDF
SchoolSize["Size Quartiles"]=pd.cut(SchoolSize["Total Students"],bins,labels=names)
SchoolSize=SchoolSize.groupby(["Size Quartiles"])
SchoolSize=SchoolSize.mean()
SchoolSize=SchoolSize[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
SchoolSize
```
